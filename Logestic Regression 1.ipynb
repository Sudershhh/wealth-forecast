```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

label_encoder = LabelEncoder()

df['country'] = label_encoder.fit_transform(df['country'])
df['country'] = df['country'].astype(float)

df['industries'] = label_encoder.fit_transform(df['industries'])
df['industries'] = df['industries'].astype(float)

df['gender'] = label_encoder.fit_transform(df['gender'])
df['gender'] = df['gender'].astype(float)

df['finalWorth'] = df['finalWorth'].astype(float)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split


def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def cost_function(X, y, theta):
    m = len(y)
    h = sigmoid(np.dot(X, theta))
    cost = (-1 / m) * (np.sum(y * np.log(h) + (1 - y) * np.log(1 - h)))
    return cost

def gradient_descent(X, y, theta, alpha, iterations):
     m = len(y)
     costs = []

     for _ in range(iterations):
         h = sigmoid(np.dot(X, theta))
         gradient = np.dot(X.T, (h - y)) / m
         theta -= alpha * gradient
         cost = cost_function(X, y, theta)
         costs.append(cost)

     return theta, costs

def predict(X, theta):
    probabilities = sigmoid(np.dot(X, theta))
    predictions = np.argmax(probabilities, axis=1)
    return predictions

# Extract features and target variable
X = df.drop('finalWorth', axis=1).values
y = df['finalWorth'].values

# Standardize the features
X = (X - np.mean(X, axis=0)) / np.std(X, axis=0)

# Add a bias term to the features
bias_column = np.ones((X.shape[0], 1))
X = np.concatenate((bias_column, X), axis=1)

# Convert target variable to one-hot encoding
num_classes = 4
y_one_hot = np.eye(num_classes)[y.astype(int)]

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_one_hot, test_size=1/3, random_state=42)

# Initialize parameters
theta_init = np.zeros((X_train.shape[1], num_classes))

# Hyperparameters
learning_rate = 0.01
num_iterations = 15000

# Train the model
theta_final, costs = gradient_descent(X_train, y_train, theta_init, learning_rate, num_iterations)

# Make predictions on the test set
y_pred = predict(X_test, theta_final)
```

```python
# Evaluate the model
accuracy = np.mean(y_pred == np.argmax(y_test, axis=1))
print(f'Training Accuracy: {accuracy * 100:.2f}%')

# Make predictions on the training set
y_train_pred = predict(X_train, theta_final)

# Evaluate the model on the training set
training_accuracy = np.mean(y_train_pred == np.argmax(y_train, axis=1))
print(f'Testing Accuracy: {training_accuracy * 100:.2f}%')
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix

# Convert predictions to class labels
y_true_labels = np.argmax(y_test, axis=1)

# Create confusion matrix
conf_matrix = confusion_matrix(y_true_labels, y_pred)

# Print confusion matrix
print("Confusion Matrix:")
print(conf_matrix)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score


# Calculate precision, recall, and F1 score
precision = precision_score(np.argmax(y_test, axis=1), y_pred, average='weighted')
recall = recall_score(np.argmax(y_test, axis=1), y_pred, average='weighted')
f1 = f1_score(np.argmax(y_test, axis=1), y_pred, average='weighted')

print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1 Score: {f1:.2f}')
```
